3: Compare and contrast Mutable and immutable ob
ects in Python with examples?

Ans:
Mutable Objects
Definition: Can be changed after creation.
In-place Modification: Changes happen in the same object.
Example:
python
Copy code
my_list = [1, 2, 3]
my_list[0] = 10
print(my_list)  # Output: [10, 2, 3]
Effect on Original Object: Modifications affect the original object.
Use Cases: Useful when you need to modify or update data in place.
Immutable Objects
Definition: Cannot be changed after creation.
In-place Modification: Modifications create a new object.
Example:
python
Copy code
my_string = "Hello"
my_string = my_string + " World"
print(my_string)  # Output: "Hello World"
Effect on Original Object: Modifications do not affect the original object; a new object is created.
Use Cases: Ideal when data should remain constant or unchangeable.

5: Explain the concept of type casting in Python with examples?

Ans:
Type Casting: It is also known as type conversion.
It is a process of converting 1 data type of a value or an object is called type conversion.
There are 2 types of conversions: Implicit, Explicit
Implicit Typecasting (also called automatic type conversion) is when Python automatically converts one data type into another without explicit instructions from the programmer. This usually happens when an operation involves different types, and Python ensures compatibility by converting the "smaller" type into the "larger" type.
Eg: a = "5"     # String
b = 3       # Integer
result = int(a) + b  
print(result)
Explicit Typecasting (also known as type conversion) is when the programmer manually converts one data type into another in Python. This is done using built-in functions like int(), float(), str(), etc. Explicit typecasting is used when you need to convert data types intentionally to avoid errors or to achieve specific results.
Eg:
Python provides built-in functions for type casting, such as int(), float(), str(), list(), tuple(), dict(), etc.
a = "5"     # String
b = 3       # Integer
result = int(a) + b  
print(result)

7:  Describe the different types of loops in Python and their use cases with examples.

ANS:
1. For Loop
•	Purpose: Iterates over a sequence (list, string, etc.).
•	Use Cases: When you know the number of iterations or want to process items in a collection.
python
Copy code
# Example
for fruit in ["apple", "banana", "cherry"]:
    print(fruit)
2. While Loop
•	Purpose: Repeats code while a condition is true.
•	Use Cases: When the number of iterations is not predetermined.
python
Copy code
# Example
count = 0
while count < 5:
    print(count)
    count += 1
3. Nested Loops
•	Purpose: A loop inside another loop.
•	Use Cases: Useful for multidimensional data structures.
python
Copy code
# Example
for row in [[1, 2, 3], [4, 5, 6]]:
    for number in row:
        print(number)
4. Loop Control Statements
•	Break: Exits the loop.
•	Continue: Skips to the next iteration.
•	Pass: Does nothing (placeholder).
python
Copy code
# Example of break
for i in range(5):
    if i == 3:
        break  # Exits loop when i is 3
    print(i)
